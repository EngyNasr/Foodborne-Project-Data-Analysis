```python
from bioblend.galaxy import GalaxyInstance
import pandas as pd
from io import StringIO
import os
```

# Add galaxy API key
Add `GALAXY_API_KEY` from `https://usegalaxy.eu/user/api_key` via `export GALAXY_API_KEY="<key>"`


```python
gi = GalaxyInstance(url='https://usegalaxy.eu/', key=os.environ['GALAXY_API_KEY'])

history = gi.histories.show_history('c6a065584d5bc954', contents=True)
```

```python
VFs_name = 'VFs of genes'
AMRs_name =  'AMR Identifiied'

summary_df = pd.DataFrame()

for dataset in history:
    if 'tags' in dataset:
        tags = '/'.join(dataset['tags'])
    else:
        tags = 'no tag'

    name = dataset['name']

    if VFs_name in name:
        print(name, tags)
        try:
            data = gi.datasets.download_dataset(dataset['id'], maxwait = 3)

            df = pd.read_csv(StringIO(data.decode("utf-8")), sep="\t")

            summary_df.loc[tags, 'VFs'] = len(df)

        except:
            pass

    if AMRs_name in name:
        print(name, tags)
        try:
            data = gi.datasets.download_dataset(dataset['id'], maxwait = 3)

            df = pd.read_csv(StringIO(data.decode("utf-8")), sep="\t")

            summary_df.loc[tags, 'AMRs'] = len(df)
        except:
            pass

print(summary_df)
```
